# Heart Disease Prediction and Risk Assessment Using Machine Learning on Clinical Health Data

## Phase 4: Machine Learning Model Development

### Objective

The objective of this phase is to develop multiple supervised machine learning models capable of predicting the presence of heart disease using the cleaned and preprocessed clinical health dataset.

In this phase, the processed training dataset will be used to train various classification algorithms, including Logistic Regression, Decision Tree, Random Forest, K-Nearest Neighbors (KNN), Support Vector Machine (SVM), and XGBoost. Each trained model will be saved for subsequent evaluation, comparison, explainability, and deployment.

This phase establishes the predictive foundation of the project, enabling the identification of the most effective machine learning model for heart disease prediction in the following phases.

## Step 4.1 Import Required Libraries

Import all the required Python libraries for loading the processed datasets, training machine learning models, and saving the trained models.

```python
import pandas as pd
import numpy as np

# Machine Learning Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# XGBoost
from xgboost import XGBClassifier

# Save Models
import joblib

print("Libraries imported successfully.")
```

## Step 4.2 Load Processed Datasets

Load the processed training and testing datasets generated during the preprocessing phase.

```python
# Unscaled datasets (Tree-based models)
X_train = pd.read_csv("../data/processed/X_train.csv")
X_test = pd.read_csv("../data/processed/X_test.csv")

# Scaled datasets (Distance-based models)
X_train_scaled = pd.read_csv("../data/processed/X_train_scaled.csv")
X_test_scaled = pd.read_csv("../data/processed/X_test_scaled.csv")

# Target variables
y_train = pd.read_csv("../data/processed/y_train.csv").squeeze()
y_test = pd.read_csv("../data/processed/y_test.csv").squeeze()

print("Datasets loaded successfully.")
```

## Step 4.3 Verify Dataset Shapes

Verify the dimensions of all training and testing datasets before training the machine learning models.

```python
print("X_train Shape        :", X_train.shape)
print("X_test Shape         :", X_test.shape)

print()

print("X_train_scaled Shape :", X_train_scaled.shape)
print("X_test_scaled Shape  :", X_test_scaled.shape)

print()

print("y_train Shape        :", y_train.shape)
print("y_test Shape         :", y_test.shape)
```

## Step 4.4 Train Logistic Regression Model

Train a Logistic Regression classifier using the scaled training dataset.

Logistic Regression is one of the most widely used baseline classification algorithms for binary classification problems. It estimates the probability of heart disease occurrence using a logistic function and serves as an important benchmark for comparing the performance of more complex machine learning models.

```python
# Initialize Logistic Regression model

logistic_model = LogisticRegression(
    random_state=42,
    max_iter=1000
)

# Train the model

logistic_model.fit(X_train_scaled, y_train)

print("Logistic Regression model trained successfully.")
```

## Step 4.5 Save Logistic Regression Model

Save the trained Logistic Regression model for future evaluation, explainability, and deployment.

```python
joblib.dump(
    logistic_model,
    "../models/logistic_regression.pkl"
)

print("Logistic Regression model saved successfully.")
```

## Step 4.6 Verify Saved Model

Verify that the Logistic Regression model has been successfully saved in the models directory.

```python
import os

os.path.exists("../models/logistic_regression.pkl")
```

## Step 4.7 Train Decision Tree Classifier

Train a Decision Tree Classifier using the unscaled training dataset.

Decision Trees classify observations by recursively splitting the data based on feature values. They are simple to interpret, capable of handling both numerical and categorical features, and can capture complex non-linear relationships within the data.

```python
# Initialize Decision Tree Classifier

decision_tree_model = DecisionTreeClassifier(
    random_state=42
)

# Train the model

decision_tree_model.fit(X_train, y_train)

print("Decision Tree model trained successfully.")
```

## Step 4.8 Save Decision Tree Model

Save the trained Decision Tree model for future evaluation, explainability, and deployment.

```python
joblib.dump(
    decision_tree_model,
    "../models/decision_tree.pkl"
)

print("Decision Tree model saved successfully.")
```

## Step 4.9 Verify Saved Model

Verify that the Decision Tree model has been successfully saved in the models directory.

```python
import os

os.path.exists("../models/decision_tree.pkl")
```

## Step 4.10 Train Random Forest Classifier

Train a Random Forest Classifier using the unscaled training dataset.

Random Forest is an ensemble learning algorithm that constructs multiple decision trees during training and combines their predictions to improve classification accuracy. It is robust to overfitting and generally provides better performance than a single decision tree.

```python
# Initialize Random Forest Classifier

random_forest_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

# Train the model

random_forest_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

## Step 4.11 Save Random Forest Model

Save the trained Random Forest model for future evaluation, explainability, and deployment.

```python
joblib.dump(
    random_forest_model,
    "../models/random_forest.pkl"
)

print("Random Forest model saved successfully.")
```

## Step 4.12 Verify Saved Model

Verify that the Random Forest model has been successfully saved in the models directory.

```python
import os

os.path.exists("../models/random_forest.pkl")
```

## Step 4.13 Train K-Nearest Neighbors (KNN) Classifier

Train a K-Nearest Neighbors (KNN) classifier using the scaled training dataset.

KNN is a distance-based supervised learning algorithm that classifies new observations based on the majority class among their nearest neighbors. Since KNN relies on distance calculations, feature scaling is essential to ensure that all variables contribute equally during prediction.

```python
# Initialize KNN Classifier

knn_model = KNeighborsClassifier(
    n_neighbors=5
)

# Train the model

knn_model.fit(X_train_scaled, y_train)

print("K-Nearest Neighbors model trained successfully.")
```

## Step 4.14 Save KNN Model

Save the trained K-Nearest Neighbors model for future evaluation, explainability, and deployment.

```python
joblib.dump(
    knn_model,
    "../models/knn.pkl"
)

print("K-Nearest Neighbors model saved successfully.")
```

## Step 4.15 Verify Saved Model

Verify that the K-Nearest Neighbors model has been successfully saved in the models directory.

```python
import os

os.path.exists("../models/knn.pkl")
```

## Step 4.16 Train Support Vector Machine (SVM)

Train a Support Vector Machine (SVM) classifier using the scaled training dataset.

Support Vector Machine is a powerful supervised learning algorithm that identifies the optimal decision boundary separating different classes. It is particularly effective for high-dimensional datasets and benefits significantly from feature scaling.

```python
# Initialize Support Vector Machine

svm_model = SVC(
    kernel="rbf",
    probability=True,
    random_state=42
)

# Train the model

svm_model.fit(X_train_scaled, y_train)

print("Support Vector Machine model trained successfully.")
```

## Step 4.17 Save SVM Model

Save the trained Support Vector Machine model for future evaluation, explainability, and deployment.

```python
joblib.dump(
    svm_model,
    "../models/svm.pkl"
)

print("Support Vector Machine model saved successfully.")
```

## Step 4.18 Verify Saved Model

Verify that the Support Vector Machine model has been successfully saved in the models directory.

```python
import os

os.path.exists("../models/svm.pkl")
```

## Step 4.19 Train XGBoost Classifier

Train an Extreme Gradient Boosting (XGBoost) classifier using the training dataset.

XGBoost is an advanced ensemble learning algorithm based on gradient boosting. It builds multiple decision trees sequentially, where each new tree corrects the errors of the previous ones. XGBoost is widely recognized for its high predictive performance, regularization capabilities, and efficiency on structured tabular datasets.

```python
# Initialize XGBoost Classifier

xgboost_model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric="logloss"
)

# Train the model

xgboost_model.fit(X_train, y_train)

print("XGBoost model trained successfully.")
```

## Step 4.20 Save XGBoost Model

Save the trained XGBoost model for future evaluation, explainability, and deployment.

```python
joblib.dump(
    xgboost_model,
    "../models/xgboost.pkl"
)

print("XGBoost model saved successfully.")
```

## Step 4.21 Verify Saved Model

Verify that the XGBoost model has been successfully saved in the models directory.

```python
import os

os.path.exists("../models/xgboost.pkl")
```

## Step 4.22 Verify All Saved Models

Verify that all trained machine learning models have been successfully saved in the `models` directory. This ensures that the models are available for evaluation, explainability, deployment, and future use without retraining.

```python
import os

model_files = [
    "logistic_regression.pkl",
    "decision_tree.pkl",
    "random_forest.pkl",
    "knn.pkl",
    "svm.pkl",
    "xgboost.pkl"
]

for model in model_files:
    print(f"{model:<30} {os.path.exists('../models/' + model)}")
```
