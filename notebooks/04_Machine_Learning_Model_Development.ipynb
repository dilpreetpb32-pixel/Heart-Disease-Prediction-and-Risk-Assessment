# Phase 4: Machine Learning Model Development

## Objective

The objective of this phase is to train multiple supervised machine learning models for predicting cardiovascular disease using the preprocessed clinical dataset.

The processed training and testing datasets prepared during the preprocessing phase are used directly to avoid repeating data preparation steps. Each model is trained independently, and prediction results are generated for evaluation in the next phase.

---

## Machine Learning Models

- Logistic Regression
- Decision Tree
- Random Forest
- Support Vector Machine (SVM)
- K-Nearest Neighbors (KNN)
- XGBoost

```python
# ==========================================================
# Step 1: Import Required Libraries
# ==========================================================

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import joblib

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score

from xgboost import XGBClassifier
```

```python
# ==========================================================
# Step 2: Load Processed Datasets
# ==========================================================

X_train = pd.read_csv("../data/processed/X_train_scaled.csv")
X_test = pd.read_csv("../data/processed/X_test_scaled.csv")

y_train = pd.read_csv("../data/processed/y_train.csv").squeeze()
y_test = pd.read_csv("../data/processed/y_test.csv").squeeze()

print("Training Features :", X_train.shape)
print("Testing Features :", X_test.shape)

print("Training Labels :", y_train.shape)
print("Testing Labels :", y_test.shape)
```

```python
# ==========================================================
# Step 3: Verify Dataset Shapes
# ==========================================================

X_train.head()
X_train.describe()
```

## Observation

The processed datasets were successfully loaded and verified.

### Key Findings

- Both scaled and unscaled datasets are available.
- The training and testing datasets contain consistent features.
- The data is ready for machine learning model development.

```python
# ==========================================================
# Step 4: Train Logistic Regression
# ==========================================================

logistic_model = LogisticRegression(
    random_state=42,
    max_iter=1000
)

logistic_model.fit(X_train, y_train)

print("Logistic Regression Model Trained Successfully.")
```

```python
# ==========================================================
# Step 5: Logistic Regression Predictions
# ==========================================================

y_pred_lr = logistic_model.predict(X_test)
```

## Observation

The Logistic Regression model was successfully trained using the scaled training dataset. Predictions were generated for the testing dataset and will be evaluated in the next phase.

```python
# ==========================================================
# Step 6: Logistic Regression Accuracy
# ==========================================================

lr_accuracy = accuracy_score(y_test, y_pred_lr)

print(f"Accuracy : {lr_accuracy:.4f}")
```

## Observation

Evaluate the prediction accuracy of the Logistic Regression model using the testing dataset.

```python
# ==========================================================
# Step 7: Saving Logistic Regression Model
# ==========================================================

joblib.dump(
    logistic_model,
    "../models/logistic_regression.pkl"
)

print("Logistic Regression Model Saved Successfully.")
```

```python
# ==========================================================
# Step 8: Train Decision Tree Model
# ==========================================================

decision_tree = DecisionTreeClassifier(
    random_state=42
)

decision_tree.fit(X_train, y_train)

print("Decision Tree Model Trained Successfully.")
```

```python
# ==========================================================
# Step 9: Decision Tree Predictions
# ==========================================================

y_pred_dt = decision_tree.predict(X_test)
```

## Observation

The Decision Tree model was successfully trained using the unscaled training dataset.

These predictions will be evaluated and compared with other machine learning models in the next phase to determine the model's overall performance.

```python
# ==========================================================
# Step 10: Decision Tree Accuracy
# ==========================================================

dt_accuracy = accuracy_score(y_test, y_pred_dt)

print(f"Accuracy : {dt_accuracy:.4f}")
```

```python
# ==========================================================
# Step 11: Saving Decison Tree Model
# ==========================================================

joblib.dump(
    decision_tree,
    "../models/decision_tree.pkl"
)

print("Decision Tree Model Saved Successfully.")
```

```python
# ==========================================================
# Step 8: Train Random Forest Model
# ==========================================================

# Initialize Random Forest Classifier
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train the model
rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

```python
# ==========================================================
# Step 9: Random Forest Predictions
# ==========================================================

# Generate predictions
y_pred_rf = rf_model.predict(X_test)

# Display first 10 predictions
print("First 10 Predictions:")
print(y_pred_rf[:10])
```

## Observation

The Random Forest model was successfully trained using the unscaled training dataset.

The model generated predictions for the testing dataset. Random Forest combines multiple decision trees to improve prediction accuracy and reduce overfitting compared to a single Decision Tree. The generated predictions will be evaluated in the next phase.

XGBoost is an ensemble boosting algorithm that builds multiple decision trees sequentially, where each new tree attempts to correct the errors of the previous ones.
