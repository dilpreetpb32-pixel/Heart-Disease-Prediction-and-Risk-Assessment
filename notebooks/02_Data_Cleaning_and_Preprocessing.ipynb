```python
# ============================================================
# Step 1: Import Required Libraries
# ============================================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

pd.set_option("display.max_columns", None)
```

```python
# ============================================================
# Step 2: Load Dataset
# ============================================================

df = pd.read_csv("../data/raw/cardio_data_processed.csv")

print("Dataset loaded successfully!\n")

display(df.head())
```

```python
# =====================================================
# Step 3: Initial Dataset Inspection
# =====================================================

print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Information:")
df.info()
```

```python
# =====================================================
# Step 4: Missing Values
# =====================================================

missing = df.isnull().sum()

display(missing)

print("\nTotal Missing Values:", missing.sum())
```

```python
# =====================================================
# Step 5: Duplicate Records
# =====================================================

duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
# =====================================================
# Step 6: Invalid Values
# =====================================================

print("Height <= 0 :", (df["height"] <= 0).sum())

print("Weight <= 0 :", (df["weight"] <= 0).sum())

print("BMI <= 0 :", (df["bmi"] <= 0).sum())

print("Systolic BP <= 0 :", (df["ap_hi"] <= 0).sum())

print("Diastolic BP <= 0 :", (df["ap_lo"] <= 0).sum())
```

```python
# =====================================================
# Step 7: Remove Unnecessary Columns
# =====================================================

columns_to_remove = [
    "id",
    "age",
    "bp_category"
]

existing_columns = [col for col in columns_to_remove if col in df.columns]

df.drop(columns=existing_columns, inplace=True)

print("Removed Columns:")
print(existing_columns)

print("\nRemaining Columns:")
print(df.columns.tolist())
```

```python
# =====================================================
# Step 8: Verify Categorical Variables
# =====================================================

categorical_features = [
    "gender",
    "cholesterol",
    "gluc",
    "smoke",
    "alco",
    "active"
]

for col in categorical_features:
    print(f"\n{col}")
    print(df[col].value_counts().sort_index())
```

```python
# =====================================================
# Step 9: Feature Scaling
# =====================================================

X = df.drop("cardio", axis=1)
y = df["cardio"]

continuous_features = [
    "age_years",
    "height",
    "weight",
    "bmi",
    "ap_hi",
    "ap_lo"
]

scaler = StandardScaler()

X[continuous_features] = scaler.fit_transform(
    X[continuous_features]
)

display(X.head())
```

```python
# =====================================================
# Step 10: Train-Test Split
# =====================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("Training Features :", X_train.shape)

print("Testing Features :", X_test.shape)

print("Training Labels :", y_train.shape)

print("Testing Labels :", y_test.shape)
```

```python
#====================================================
# Saving the cleaned dataset to a CSV file
#====================================================

processed_df = pd.concat([X, y], axis=1)

processed_df.to_csv(
    "../data/processed/cardio_data_cleaned.csv",
    index=False
)

print("Cleaned dataset saved successfully!")
```

```python
#====================================================
# Savin train and test datasets to CSV files
#====================================================
X_train.to_csv("../data/processed/X_train.csv", index=False)

X_test.to_csv("../data/processed/X_test.csv", index=False)

y_train.to_csv("../data/processed/y_train.csv", index=False)

y_test.to_csv("../data/processed/y_test.csv", index=False)

print("Training and testing datasets saved successfully!")
```

# Phase 2: Data Cleaning & Preprocessing

## Objective

The objective of this phase was to improve the quality of the dataset and prepare it for machine learning by checking for missing values, duplicate records, and invalid entries, removing unnecessary features, verifying categorical variables, applying feature scaling, and splitting the dataset into training and testing sets.

---

## Data Cleaning Summary

### Missing Values

- All columns were checked for missing values.
- No missing values were found; therefore, no imputation was required.

### Duplicate Records

- The dataset was examined for duplicate records.
- Duplicate rows were removed if present to avoid bias during model training.

### Invalid Values

The dataset was validated for unrealistic or inconsistent values in:

- Height
- Weight
- Body Mass Index (BMI)
- Systolic Blood Pressure (`ap_hi`)
- Diastolic Blood Pressure (`ap_lo`)

Additionally, records where systolic blood pressure was lower than diastolic blood pressure were identified for further review.

### Removed Columns

The following columns were removed before model training:

- **id**: Unique identifier with no predictive value.
- **age**: Removed because **age_years** contains the same information in a more interpretable format.
- **bp_category**: Removed because it is derived from blood pressure measurements (`ap_hi` and `ap_lo`) and introduces redundant information.

### Categorical Variables

All categorical variables were already encoded as numeric values, so no further encoding was required.

### Feature Scaling

Continuous numerical features were standardized using **StandardScaler** to ensure that all variables contribute equally during model training.

### Train-Test Split

The dataset was divided into:

- **80% Training Data**
- **20% Testing Data**

Stratified sampling was used to preserve the class distribution in both subsets.

---

## Conclusion

The dataset has been successfully cleaned and preprocessed. Unnecessary columns were removed, data quality was verified, numerical features were standardized, and the dataset was split into training and testing sets. The processed data is now ready for exploratory data analysis and machine learning model development.
