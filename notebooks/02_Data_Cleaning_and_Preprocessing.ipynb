```python
# =====================================================
# Step 1: Import Required Libraries
# =====================================================

import os
import warnings

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)

print("Libraries imported successfully.")
```

```python
# =====================================================
# Step 2: Load Dataset
# =====================================================

df = pd.read_csv("../data/raw/cardio_data_processed.csv")

print("Dataset Loaded Successfully!")

display(df.head())
```

```python
# =====================================================
# Step 3: Dataset Overview
# =====================================================

print("Shape:", df.shape)

print("\nColumns")

display(df.columns)

print("\nData Types")

display(df.dtypes)

print("\nDataset Information")

df.info()
```

```python
# =====================================================
# Step 4: Missing Values
# =====================================================

missing = df.isnull().sum()

display(missing)

print("\nTotal Missing Values:", missing.sum())
```

```python
# =====================================================
# Step 5: Duplicate Records
# =====================================================

duplicates = df.duplicated().sum()

print(f"Duplicate Rows (before removing identifier columns): {duplicates}")
```

```python
# =====================================================
# Step 6: Verify Data Types
# =====================================================

display(df.dtypes)
```

```python
# =====================================================
# Step 7: Validate Numerical Features
# =====================================================

numerical_features = [
    "age",
    "age_years",
    "height",
    "weight",
    "bmi",
    "ap_hi",
    "ap_lo"
]

display(df[numerical_features].describe())
```

```python
# =====================================================
# Step 8: Verify Categorical Features
# =====================================================

categorical_features = [
    "gender",
    "cholesterol",
    "gluc",
    "smoke",
    "alco",
    "active"
]

for column in categorical_features:
    print(f"\n{column}")
    print(df[column].value_counts())
```

```python
# =====================================================
# Step 9: Remove Unnecessary Columns
# =====================================================

columns_to_drop = [
    "id",
    "age",
    "bp_category_encoded",
    "bp_category"
]

df.drop(
    columns=columns_to_drop,
    errors="ignore",
    inplace=True
)

print("Columns removed successfully.")

display(df.head())
```

```python
# =====================================================
# Step 10: Separate Features and Target
# =====================================================

X = df.drop("cardio", axis=1)

y = df["cardio"]

print("Feature Matrix Shape :", X.shape)

print("Target Shape :", y.shape)
```

```python
# =====================================================
# Step 11: Train-Test Split
# =====================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Features :", X_train.shape)
print("Testing Features :", X_test.shape)

print("Training Labels :", y_train.shape)
print("Testing Labels :", y_test.shape)
```

```python
# =====================================================
# Step 12: Feature Scaling
# =====================================================

from sklearn.preprocessing import StandardScaler

# Numerical features to scale
continuous_features = [
    "age_years",
    "height",
    "weight",
    "bmi",
    "ap_hi",
    "ap_lo"
]

# Create copies so original data remains unchanged
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# Initialize scaler
scaler = StandardScaler()

# Scale only numerical features
X_train_scaled[continuous_features] = scaler.fit_transform(
    X_train[continuous_features]
)

X_test_scaled[continuous_features] = scaler.transform(
    X_test[continuous_features]
)

print("Feature scaling completed successfully.")
```

```python
# ==========================================================
# Save Processed Datasets
# ==========================================================

X_train.to_csv("../data/processed/X_train.csv", index=False)
X_test.to_csv("../data/processed/X_test.csv", index=False)

X_train_scaled.to_csv("../data/processed/X_train_scaled.csv", index=False)
X_test_scaled.to_csv("../data/processed/X_test_scaled.csv", index=False)

y_train.to_csv("../data/processed/y_train.csv", index=False)
y_test.to_csv("../data/processed/y_test.csv", index=False)

print("All processed datasets saved successfully.")
```

# Phase 2: Data Cleaning & Preprocessing

## Objective

The objective of this phase is to improve the quality of the dataset and prepare it for machine learning by identifying data quality issues, validating feature values, removing unnecessary attributes, splitting the data into training and testing sets, and standardizing numerical features.

---

## Tasks Performed

### Missing Values

The dataset was examined for missing values. No missing values were detected; therefore, no imputation was required.

### Duplicate Records

Duplicate records were checked before removing identifier columns. No duplicate records requiring removal were found. During later analysis, identical feature values may appear because different patients can share the same clinical measurements after removing the unique identifier (`id`).

### Data Validation

The numerical variables were reviewed to verify that values were within realistic clinical ranges. Features such as age, height, weight, BMI, systolic blood pressure, and diastolic blood pressure were inspected for inconsistencies.

### Categorical Features

All categorical variables were verified and were already encoded into numerical values suitable for machine learning algorithms.

### Feature Selection

The following columns were removed before model training:

- **id** – Unique patient identifier with no predictive value.
- **age** – Removed because `age_years` provides the same information in a more interpretable format.
- **bp_category_encoded** – Removed because it is derived from systolic and diastolic blood pressure (`ap_hi` and `ap_lo`) and does not provide additional independent information.

### Train-Test Split

The cleaned dataset was divided into training and testing subsets using an 80:20 ratio. Stratified sampling was applied to preserve the class distribution of the target variable.

### Feature Scaling

Continuous numerical variables were standardized using `StandardScaler`. The scaler was fitted only on the training data and then applied to both the training and testing datasets to prevent data leakage.

---

## Output

- Cleaned dataset
- Processed dataset ready for machine learning
- Training feature set (`X_train`)
- Testing feature set (`X_test`)
- Training target set (`y_train`)
- Testing target set (`y_test`)

The dataset has been successfully cleaned and prepared for machine learning.
