# Phase 2: Data Cleaning and Preprocessing

## Objective

The objective of this phase is to clean and preprocess the cardiovascular disease dataset before developing machine learning models.

The preprocessing steps include removing unnecessary features, preparing the input and target variables, splitting the dataset into training and testing sets, scaling numerical features, and saving the processed datasets for subsequent phases.

Proper preprocessing ensures consistency throughout model training, evaluation, optimization, and deployment.

## Step 2.1 Import Required Libraries

Import the required Python libraries for data preprocessing, dataset splitting, feature scaling, and saving processed files.

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import joblib

pd.set_option("display.max_columns", None)
```

## Step 2.2 Load the Dataset

Load the processed cardiovascular disease dataset obtained from the previous phase.

```python
df = pd.read_csv("../data/raw/cardio_data_processed.csv")

print("Dataset Loaded Successfully")
```

## Step 2.3 Display Dataset

Display the first five observations to verify that the dataset has been loaded correctly.

```python
df.head()
```

## Step 2.4 Check Dataset Dimensions

Display the number of rows and columns present in the dataset.

```python
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

## Step 2.5 Remove Unnecessary Features

Remove columns that do not contribute to model learning or contain redundant information.

```python
df.drop(
    columns=[
        "id",
        "age",
        "bp_category",
        "bp_category_encoded"
    ],
    inplace=True
)

print("Columns removed successfully.")

df.columns.tolist()
```

## Step 2.6 Check Missing Values

Verify that the cleaned dataset does not contain missing values.

```python
df.isnull().sum()
```

## Step 2.7 Check Duplicate Records

Verify that there are no duplicate records remaining in the cleaned dataset.

```python
print("Duplicate Rows :", df.duplicated().sum())
```

## Step 2.8 Define Features and Target Variable

Separate the predictor variables (X) from the target variable (y).

The target variable is **cardio**, which indicates the presence or absence of heart disease.

```python
X = df.drop("cardio", axis=1)

y = df["cardio"]

print(X.shape)
print(y.shape)
```

## Step 2.9 Remove Duplicate Records

After removing unnecessary features, some observations become identical across all remaining variables. These duplicate records are removed to avoid redundant information during model training.

```python
print("Shape Before:", df.shape)

df = df.drop_duplicates()

print("Shape After:", df.shape)

print("Duplicate Rows:", df.duplicated().sum())
```

## Step 2.10 Define Features and Target Variable

After removing duplicate records, redefine the predictor variables (**X**) and the target variable (**y**) using the cleaned dataset.

- **X** contains all the input features used for model training.
- **y** contains the target variable (`cardio`), which indicates whether a patient has heart disease.

This ensures that the machine learning models are trained using the final cleaned dataset without any redundant observations.

```python
X = df.drop("cardio", axis=1)
y = df["cardio"]

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

## Step 2.11 Split the Dataset into Training and Testing Sets

Split the cleaned dataset into training and testing subsets.

- **Training Set (80%)**: Used to train the machine learning models.
- **Testing Set (20%)**: Used to evaluate the performance of the trained models on unseen data.

A fixed `random_state` is used to ensure reproducibility, while `stratify=y` preserves the class distribution in both training and testing datasets.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)

print("Training Target   :", y_train.shape)
print("Testing Target    :", y_test.shape)
```

## Step 2.12 Standardize Numerical Features

Standardize the numerical features using the **StandardScaler**.

The scaler is fitted only on the training data to prevent data leakage. The same scaler is then applied to both the training and testing datasets.

Only continuous numerical features are standardized, while categorical features remain unchanged.

```python
numerical_features = [
    "age_years",
    "height",
    "weight",
    "bmi",
    "ap_hi",
    "ap_lo"
]

scaler = StandardScaler()

X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numerical_features] = scaler.fit_transform(
    X_train[numerical_features]
)

X_test_scaled[numerical_features] = scaler.transform(
    X_test[numerical_features]
)

print("Feature Scaling Completed Successfully.")
```

## Step 2.13 Save the Processed Datasets

Save the cleaned and preprocessed datasets for reuse in subsequent phases of the project.

The processed datasets are stored separately to ensure reproducibility and eliminate the need to repeat preprocessing during model development and evaluation.

```python
X_train_scaled.to_csv(
    "../data/processed/X_train_scaled.csv",
    index=False
)

X_test_scaled.to_csv(
    "../data/processed/X_test_scaled.csv",
    index=False
)

y_train.to_frame(name="cardio").to_csv(
    "../data/processed/y_train.csv",
    index=False
)

y_test.to_frame(name="cardio").to_csv(
    "../data/processed/y_test.csv",
    index=False
)

df.to_csv(
    "../data/processed/cardio_data_cleaned.csv",
    index=False
)

print("Processed datasets saved successfully.")
```

## Step 2.14 Save the StandardScaler

Save the fitted StandardScaler object for future use.

The same scaler will later be used to transform user input in the deployed Streamlit web application, ensuring consistency between training and prediction.

```python
joblib.dump(
    scaler,
    "../models/scaler.pkl"
)

print("StandardScaler saved successfully.")
```

## Step 2.15 Verify the Saved Files

Verify that all processed datasets and the fitted scaler have been successfully saved.

```python
import os

files = [
    "../data/processed/cardio_data_cleaned.csv",
    "../data/processed/X_train_scaled.csv",
    "../data/processed/X_test_scaled.csv",
    "../data/processed/y_train.csv",
    "../data/processed/y_test.csv",
    "../models/scaler.pkl"
]

for file in files:
    print(
        f"{os.path.basename(file)} :",
        os.path.exists(file)
    )
```
