```python
# ============================================================
# Step 1: Import Required Libraries
# ============================================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

pd.set_option("display.max_columns", None)
```

```python
# ============================================================
# Step 2: Load Dataset
# ============================================================

df = pd.read_csv("../data/raw/cardio_data_processed.csv")

print("Dataset loaded successfully!\n")

display(df.head())
```

```python
# =====================================================
# Step 3: Initial Dataset Inspection
# =====================================================

print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Information:")
df.info()
```

```python
# =====================================================
# Step 4: Missing Values
# =====================================================

missing = df.isnull().sum()

display(missing)

print("\nTotal Missing Values:", missing.sum())
```

```python
# =====================================================
# Step 5: Duplicate Records
# =====================================================

duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
# =====================================================
# Step 6: Invalid Values
# =====================================================

print("Height <= 0 :", (df["height"] <= 0).sum())

print("Weight <= 0 :", (df["weight"] <= 0).sum())

print("BMI <= 0 :", (df["bmi"] <= 0).sum())

print("Systolic BP <= 0 :", (df["ap_hi"] <= 0).sum())

print("Diastolic BP <= 0 :", (df["ap_lo"] <= 0).sum())
```

```python
# =====================================================
# Step 7: Remove Unnecessary Columns
# =====================================================

columns_to_remove = [
    "id",
    "age",
    "bp_category"
]

existing_columns = [col for col in columns_to_remove if col in df.columns]

df.drop(columns=existing_columns, inplace=True)

print("Removed Columns:")
print(existing_columns)

print("\nRemaining Columns:")
print(df.columns.tolist())
```

```python
# =====================================================
# Step 8: Verify Categorical Variables
# =====================================================

categorical_features = [
    "gender",
    "cholesterol",
    "gluc",
    "smoke",
    "alco",
    "active"
]

for col in categorical_features:
    print(f"\n{col}")
    print(df[col].value_counts().sort_index())
```

```python
# =====================================================
# Step 9: Split Dataset into Training and Testing Sets
# =====================================================

# Separate features and target variable
X = df.drop("cardio", axis=1)
y = df["cardio"]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Dataset successfully split.\n")

print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)
print("Training Labels   :", y_train.shape)
print("Testing Labels    :", y_test.shape)
```

```python
# =====================================================
# Step 10: Feature Scaling
# =====================================================

continuous_features = [
    "age_years",
    "height",
    "weight",
    "bmi",
    "ap_hi",
    "ap_lo"
]

# Initialize StandardScaler
scaler = StandardScaler()

# Fit ONLY on training data
X_train[continuous_features] = scaler.fit_transform(
    X_train[continuous_features]
)

# Use the same scaler on testing data
X_test[continuous_features] = scaler.transform(
    X_test[continuous_features]
)

print("Feature scaling completed successfully.\n")

print("Scaled Training Data:")
display(X_train.head())

print("Scaled Testing Data:")
display(X_test.head())
```

```python
# =====================================================
# Step 11: Save Processed Dataset
# =====================================================

import os

# Create processed folder if it doesn't exist
os.makedirs("../data/processed", exist_ok=True)

# Save complete cleaned dataset (before splitting)
df.to_csv("../data/processed/cardio_data_cleaned.csv", index=False)

# Save training and testing datasets
X_train.to_csv("../data/processed/X_train.csv", index=False)
X_test.to_csv("../data/processed/X_test.csv", index=False)

y_train.to_csv("../data/processed/y_train.csv", index=False)
y_test.to_csv("../data/processed/y_test.csv", index=False)

print("Processed datasets saved successfully!")
```

# Phase 2: Data Cleaning & Preprocessing

## Objective

The objective of this phase is to prepare the dataset for machine learning by improving its quality, identifying and handling data quality issues, removing unnecessary features, validating the dataset, scaling numerical variables, and splitting the data into training and testing sets.

---

# Data Cleaning & Preprocessing Summary

During this phase, the dataset was thoroughly examined and prepared to ensure it is suitable for machine learning model development.

### 1. Missing Values

- All columns were checked for missing values.
- No missing values were found in the dataset.
- Therefore, no imputation or missing value treatment was required.

### 2. Duplicate Records

- The dataset was checked for duplicate observations.
- Duplicate rows were removed if present to prevent bias during model training and evaluation.

### 3. Invalid or Inconsistent Values

The dataset was validated for unrealistic or inconsistent values in important clinical features, including:

- Height
- Weight
- Body Mass Index (BMI)
- Systolic Blood Pressure (`ap_hi`)
- Diastolic Blood Pressure (`ap_lo`)

Additionally, records where systolic blood pressure was lower than diastolic blood pressure were identified for further inspection.

### 4. Feature Selection

The following columns were removed before model training:

- **id** – A unique identifier that does not contribute to prediction.
- **age** – Removed because **age_years** contains the same information in a more interpretable format.
- **bp_category** – Removed because it is derived from blood pressure measurements (`ap_hi` and `ap_lo`) and introduces redundant information.

Removing these columns simplifies the dataset while preserving all relevant predictive information.

### 5. Categorical Variables

The categorical variables were reviewed to ensure they were already encoded in numerical form.

The following features required no additional encoding:

- gender
- cholesterol
- gluc
- smoke
- alco
- active

### 6. Train-Test Split

The cleaned dataset was divided into training and testing sets using an **80:20 ratio**.

Stratified sampling was applied to maintain the original distribution of the target variable (`cardio`) in both subsets, ensuring fair model training and evaluation.

### 7. Feature Scaling

Continuous numerical features were standardized using **StandardScaler**.

To prevent **data leakage**, the scaler was fitted only on the training dataset and then used to transform both the training and testing datasets.

The following numerical features were scaled:

- age_years
- height
- weight
- bmi
- ap_hi
- ap_lo

---

# Output of This Phase

At the end of this phase, the following outputs were generated:

- Cleaned dataset
- Processed dataset ready for machine learning
- Training feature set (`X_train`)
- Testing feature set (`X_test`)
- Training target set (`y_train`)
- Testing target set (`y_test`)

# Conclusion

The dataset has been successfully cleaned and preprocessed for machine learning. Data quality issues were verified, unnecessary columns were removed, categorical variables were validated, numerical features were standardized, and the dataset was divided into training and testing sets using industry-standard practices. The processed dataset is now ready for exploratory data analysis and machine learning model development.
