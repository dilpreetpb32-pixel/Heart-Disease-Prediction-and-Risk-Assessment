```python
feature_dict = {
    'id': 'Unique identifier for each patient (drop before modeling — not predictive)',
    'age': 'Age of patient in days (raw form from original data source)',
    'age_years': 'Age of patient in years (derived from age; more interpretable)',
    'gender': 'Gender: 1 = one category, 2 = other (verify mapping with data source docs)',
    'height': 'Height in cm',
    'weight': 'Weight in kg',
    'bmi': 'Body Mass Index, derived as weight(kg) / height(m)^2 — indicator of body fat',
    'ap_hi': 'Systolic blood pressure (higher/top number)',
    'ap_lo': 'Diastolic blood pressure (lower/bottom number)',
    'cholesterol': 'Cholesterol level: 1 = normal, 2 = above normal, 3 = well above normal',
    'gluc': 'Glucose level: 1 = normal, 2 = above normal, 3 = well above normal',
    'smoke': 'Whether patient smokes: 0 = No, 1 = Yes',
    'alco': 'Whether patient consumes alcohol: 0 = No, 1 = Yes',
    'active': 'Whether patient is physically active: 0 = No, 1 = Yes',
    'cardio': 'TARGET: Presence of cardiovascular disease: 0 = No, 1 = Yes'
}

for feature, description in feature_dict.items():
    print(f"{feature:12s} -> {description}")
```

```python
# Sanity check: recompute BMI and compare to existing column
df['bmi_check'] = df['weight'] / ((df['height'] / 100) ** 2)

# Compare
comparison = (df['bmi'] - df['bmi_check']).abs()
print("Max difference between given bmi and recomputed bmi:", comparison.max())
print("Rows with mismatch > 0.1:", (comparison > 0.1).sum())

df.drop(columns=['bmi_check'], inplace=True)
```

```python
# Missing values
print("Missing values per column:")
print(df.isnull().sum())

# Duplicate rows
print(f"\nDuplicate rows: {df.duplicated().sum()}")

# Convert age from days to years for easier understanding (exploration only)
df['age_years'] = (df['age'] / 365).round().astype(int)
print(f"\nAge range: {df['age_years'].min()} to {df['age_years'].max()} years")
```

# Dataset Validation

To ensure the quality and consistency of the dataset, several validation checks were performed.

### BMI Verification

The BMI value provided in the dataset was recalculated using the standard formula:

**BMI = Weight (kg) / Height² (m²)**

The calculated BMI values were compared with the existing BMI column to verify their accuracy.

---

### Missing Value Analysis

Every column was examined for missing values.

This helps identify whether data cleaning or imputation will be required before model training.

---

### Duplicate Record Analysis

The dataset was checked for duplicate patient records.

Duplicate records, if present, can negatively affect model performance and may need to be removed during preprocessing.

---

### Age Conversion

The original dataset stores patient age in **days**.

For easier interpretation, an additional feature named **age_years** was created by converting age from days into years.

This improves readability during analysis while preserving the original information.
